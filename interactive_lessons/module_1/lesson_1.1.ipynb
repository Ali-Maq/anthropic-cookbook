## ðŸ¤” What is an API?

Think of an API (Application Programming Interface) as a **waiter at a restaurant**:

1. **You (your code)** = The customer
2. **The API** = The waiter who takes your order
3. **Claude (the AI)** = The kitchen that prepares your food
4. **The response** = Your delicious meal!

You tell the waiter what you want â†’ The waiter brings your request to the kitchen â†’ The kitchen prepares it â†’ The waiter brings it back to you.

### Why Use an API?

Instead of using Claude in a web browser, the API lets you:
- Integrate Claude into YOUR applications
- Automate tasks (no clicking buttons!)
- Process hundreds or thousands of requests
- Build custom tools and features

## ðŸ”‘ Step 1: Set Up Your Environment

First, let's install the Anthropic Python library and load our API key.

````python
# Install required packages (run this once)
# Uncomment the line below if you haven't installed these packages yet
# !pip install anthropic python-dotenv

# LINE-BY-LINE EXPLANATION:
# ---------------------------
# pip install = Python's package installer command
# anthropic = The official Anthropic Python SDK (Software Development Kit)
# python-dotenv = A package to load environment variables from .env files (keeps your API key secret!)
````

````python
# Import the libraries we need
import os  # LINE 1: 'os' is Python's operating system interface - lets us read environment variables
from dotenv import load_dotenv  # LINE 2: 'load_dotenv' reads your .env file and loads API keys
from anthropic import Anthropic  # LINE 3: Import the main Anthropic class - this is what talks to Claude!

# Load your API key from .env file
load_dotenv()  # LINE 5: This function reads the .env file in your directory

# Create an Anthropic client
client = Anthropic(
    api_key=os.environ.get("ANTHROPIC_API_KEY")  # LINE 8-9: Get API key from environment variables
)

# WHAT JUST HAPPENED?
# --------------------
# 1. We imported necessary tools
# 2. We loaded environment variables (including your secret API key)
# 3. We created a 'client' object - think of this as your phone to call Claude!
#
# The 'client' is now ready to send messages to Claude and receive responses.
````

### ðŸ” Understanding the Client

The `client` object is your **interface** to Claude. Every time you want to:
- Ask a question
- Generate text
- Analyze an image
- Use a tool

...you'll use this `client` object to do it.

---

## ðŸ’¬ Step 2: Your First API Call

Let's send a simple message to Claude and get a response!

````python
# Make your first API call!
response = client.messages.create(
    model="claude-3-5-sonnet-20241022",
    max_tokens=1024,
    messages=[
        {"role": "user", "content": "Hello, Claude! Can you explain what you are in one sentence?"}
    ]
)

# LINE-BY-LINE EXPLANATION:
# ---------------------------
#
# LINE 1: response = client.messages.create(
#   - 'response' is a variable that will store Claude's answer
#   - 'client.messages.create()' is the method to send a message to Claude
#   - We're calling a function that will return Claude's response
#
# LINE 2: model="claude-3-5-sonnet-20241022",
#   - This specifies WHICH version of Claude to use
#   - "claude-3-5-sonnet-20241022" is the latest, fastest, and smartest model
#   - Different models have different capabilities and costs
#
# LINE 3: max_tokens=1024,
#   - 'max_tokens' limits how long the response can be
#   - 1 token â‰ˆ 4 characters or Â¾ of a word
#   - 1024 tokens â‰ˆ 750-800 words (a good length for most answers)
#   - This prevents unexpectedly long (and expensive!) responses
#
# LINE 4-6: messages=[...]
#   - 'messages' is a LIST (array) of message objects
#   - Each message is a dictionary with 'role' and 'content'
#
# LINE 5: {"role": "user", "content": "Hello, Claude!..."}
#   - 'role': "user" means YOU are saying this (not Claude)
#   - 'content': The actual text of your message
#   - This is like saying "I (the user) am asking: Hello, Claude!..."

print("API call successful! Response received.")
print("\nLet's examine what we got back...")
````

### ðŸŽ¯ What Just Happened?

1. We called `client.messages.create()` - this sends an HTTP request to Anthropic's servers
2. Our message traveled over the internet to Claude
3. Claude processed our question
4. Claude sent back a response
5. We stored that response in the `response` variable

---

## ðŸ” Step 3: Understanding the Response

Let's look at what Claude sent back to us.

````python
# Let's see the ENTIRE response object
print("Full Response Object:")
print("=" * 50)
print(response)
print("=" * 50)

# WHAT YOU'RE SEEING:
# -------------------
# The response is a complex object with several parts:
#
# - id: A unique identifier for this conversation turn
# - type: Always "message" for these API calls
# - role: "assistant" (meaning Claude is responding)
# - content: The actual answer (this is what we care about!)
# - model: Which Claude model was used
# - stop_reason: Why Claude stopped generating (usually "end_turn")
# - usage: How many tokens were used (important for billing!)
````

````python
# Let's examine each part of the response

print("Response ID:")
print(response.id)
print("This is a unique identifier for this specific message.\n")

print("Model Used:")
print(response.model)
print("This confirms which Claude model processed your request.\n")

print("Role:")
print(response.role)
print("This tells us who's responding - 'assistant' means Claude.\n")

print("Stop Reason:")
print(response.stop_reason)
print("Why did Claude stop? Usually 'end_turn' means it finished naturally.\n")

print("Token Usage:")
print(f"Input tokens: {response.usage.input_tokens}")
print(f"Output tokens: {response.usage.output_tokens}")
print("This shows how many tokens were used (important for cost tracking!).\n")
````

### ðŸ’° Understanding Token Usage

- **Input tokens**: Your question to Claude (you pay for this)
- **Output tokens**: Claude's response (you pay for this too)
- **Total cost** = (input tokens Ã— input price) + (output tokens Ã— output price)

For Claude 3.5 Sonnet:
- Input: $3 per million tokens
- Output: $15 per million tokens

---

## ðŸ“ Step 4: Extracting Claude's Answer

The actual text response is nested inside the `content` field.

````python
# The content is a LIST of content blocks
print("Content structure:")
print(response.content)
print()

# LINE-BY-LINE EXPLANATION:
# ---------------------------
# The 'content' is a list because Claude can return multiple types of content:
# - Text responses
# - Tool use requests
# - Thinking blocks (in extended thinking mode)
#
# Usually, it's just one text block like: [ContentBlock(text="...", type="text")]
````

````python
# Extract the text from the first content block
answer = response.content[0].text

# LINE-BY-LINE EXPLANATION:
# ---------------------------
# response.content = The list of content blocks
# [0] = Get the first item in the list (Python uses 0-based indexing)
# .text = Extract just the text string from that content block

print("\n" + "=" * 50)
print("CLAUDE'S ANSWER:")
print("=" * 50)
print(answer)
print("=" * 50)
````

## ðŸŽ¯ Practice Exercise 3: Add Error Handling

**Task**: Make your function safer by handling errors.

````python
# Improved version with error handling

def ask_claude_safe(question):
    """
    Ask Claude a question with error handling.
    
    Args:
        question (str): The question to ask
    
    Returns:
        str: Claude's answer or an error message
    """
    try:
        # TRY to make the API call
        response = client.messages.create(
            model="claude-3-5-sonnet-20241022",
            max_tokens=1024,
            messages=[
                {"role": "user", "content": question}
            ]
        )
        return response.content[0].text
    
    except Exception as e:
        # If anything goes wrong, return an error message
        return f"Error: {str(e)}"

# LINE-BY-LINE EXPLANATION:
# ---------------------------
# try: = Attempt to run this code
# except Exception as e: = If ANY error occurs, catch it
# str(e) = Convert the error to a readable string
#
# This prevents your program from crashing if:
# - Your internet is down
# - Your API key is invalid
# - Anthropic's servers are having issues
# - You hit rate limits

# Test it!
result = ask_claude_safe("What is the capital of France?")
print(result)
````

## ðŸŽ¯ Challenge: Build a Simple Chatbot

**Goal**: Create a command-line chatbot that keeps asking questions until you type 'quit'.

````python
# CHALLENGE SOLUTION

def simple_chatbot():
    """
    A simple chatbot that answers questions until you quit.
    """
    print("Simple Claude Chatbot")
    print("Type 'quit' to exit\n")
    
    while True:
        # Get user input
        question = input("You: ")
        
        # Check if user wants to quit
        if question.lower() == 'quit':
            print("Goodbye!")
            break
        
        # Ask Claude
        response = client.messages.create(
            model="claude-3-5-sonnet-20241022",
            max_tokens=1024,
            messages=[
                {"role": "user", "content": question}
            ]
        )
        
        # Print Claude's response
        answer = response.content[0].text
        print(f"\nClaude: {answer}\n")

# Uncomment to run (won't work well in Jupyter, better in a .py file)
# simple_chatbot()
````
